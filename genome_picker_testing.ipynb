```python
def build_reference_dict(annotation_dir):
 
    # Initialize the reference dictionary
    reference_dict = {}

    # List all genome directories (excluding irrelevant entries)
    genome_dirs = [d for d in os.listdir(annotation_dir) if os.path.isdir(os.path.join(annotation_dir, d))]

    # Loop through each genome directory
    for genome in genome_dirs:
        genome_dir = os.path.join(annotation_dir, genome)

        # Find files based on their patterns
        gtf_ref = glob.glob(os.path.join(genome_dir, "*.gtf"))
        ref_fa = glob.glob(os.path.join(genome_dir, "*.fa"))
        refflat_ref = glob.glob(os.path.join(genome_dir, "*refflat.txt"))
        ribo_ref = glob.glob(os.path.join(genome_dir, "*ribosomalIntervalsWheader.txt"))
        chromap_index = glob.glob(os.path.join(genome_dir, "*chromap"))
        snps_bed = glob.glob(os.path.join(genome_dir, "*NGSCheckMate_andInterestingSNP.bed"))

        # Construct the dictionary for this genome by type
        reference_dict[genome] = {
            "star_align": os.path.join(genome_dir, "star") if os.path.exists(os.path.join(genome_dir, "star")) else None,
            "htseq_count": gtf_ref[0] if gtf_ref else None,
            "picard_metrics": {
                "fa": ref_fa[0] if ref_fa else None,
                "flat": refflat_ref[0] if refflat_ref else None,
                "ribosomal_intervals": ribo_ref[0] if ribo_ref else None,
            },
            "snps": {
                "fa": ref_fa[0] if ref_fa else None,
                "snps_bed": snps_bed[0] if snps_bed else None,
            } if "GRCh" in genome or "hg" in genome else None,  # SNPs only for human genomes
            "chromap": {
                "fa": ref_fa[0] if ref_fa else None,
                "index": chromap_index[0] if chromap_index else None,
            },
            "macs": "hs" if "GRCh" in genome or "hg" in genome else "mm" if "GRCm" in genome or "mm" in genome else None,
            "bigwigs": "2913022398" if "GRCh" in genome or "hg" in genome else "2652783500" if "GRCm" in genome or "mm" in genome else None,
        }

    return reference_dict
```

```python
reference_dict = build_reference_dict(ANNOTATIONS_DIR)
```

```python
def genome_picker_v3(type, sample):
    # Get genome for the current sample
    try:
        genome = sample_sheet["genome"].loc[sample_sheet["sample_flowcell"] == sample].item()
    except ValueError:
        genome = sample_sheet["genome"].loc[sample_sheet["sample"] == sample].item()

    reference = reference_dict[genome][type]

    return reference
```

```python
def genome_picker_v2(type, sample):
    # Get genome for the current sample
    try:
        genome = sample_sheet["genome"].loc[sample_sheet["sample_flowcell"] == sample].item()
    except ValueError:
        genome = sample_sheet["genome"].loc[sample_sheet["sample"] == sample].item()

    # Define the base directory for annotations
    genome_dir = os.path.join(ANNOTATIONS_DIR, genome)
    
    if not os.path.exists(genome_dir):
        raise Exception(f"Genome folder for {genome} does not exist in {ANNOTATIONS_DIR}")

    # Helper function to detect files
    def find_file(pattern):
        files = glob.glob(os.path.join(genome_dir, pattern))
        if not files:
            #logger.debug(f"File matching pattern '{pattern}' not found in {genome_dir}")
            return None
        return files[0]  # Return the first match

    # Locate required files
    gtf_ref = find_file("*.gtf")
    ref_fa = find_file("*.fa")
    refflat_ref = find_file("*refflat.txt")
    ribo_ref = find_file("*ribosomalIntervalsWheader.txt")
    chromap_index = find_file("*chromap")

    # Match type to the appropriate reference
    if type == "star_align":
        reference = os.path.join(genome_dir, "star")

    elif type == "htseq_count":
        reference = gtf_ref

    elif type == "picard_metrics":
        reference = {
            "fa": ref_fa,
            "flat": refflat_ref,
            "ribosomal_intervals": ribo_ref,
        }

    elif type == "snps":
        try:
        # SNPs can only be done for human genomes
            if "GRCh" not in genome and "hg" not in genome:
                raise Exception("SNPs are currently supported only for human genomes (e.g., GRCh38 or hg19).")

            snps_bed = find_file("*NGSCheckMate_andInterestingSNP.bed")
            reference = {
            "fa": ref_fa,
            "snps_bed": snps_bed,
            }
        except Exception as e:
            raise Exception(f"Error in locating SNP references for genome '{genome}': {str(e)}")



    elif type == "chromap":
        reference = {
            "fa": ref_fa,
            "index": chromap_index,
        }

    elif type == "macs":
        if "GRCh" in genome or "hg" in genome:
            reference = "hs"
        elif "GRCm" in genome or "mm" in genome:
            reference = "mm"
        else:
            raise Exception("Support for custom genome sizes not implemented yet.")

    elif type == "bigwigs":
        if "GRCh" in genome or "hg" in genome:
            reference = "2913022398"
        elif "GRCm" in genome or "mm" in genome:
            reference = "2652783500"
        else:
            raise Exception("Support for custom genome sizes not implemented yet.")

    return reference
```
